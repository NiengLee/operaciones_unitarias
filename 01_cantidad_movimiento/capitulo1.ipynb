# Numero de Reynolds
$$
Re =  \frac{\rho \vec{V} l}{\mu}
$$

- $\rho$ es la densidad del fluido
- $\vec{V}$ es la velocidad del fluido
- $l$ es la longitud caracteristica del fluido, longitud en la cual existe un gradiente de velocidad.
- $\mu$ es la viscosidad del fluido

- A bajo Re flujo laminar
- A alto Re flujo turbulento

## Principios Fundamentales
### En tubos, l es el diámetro
- Re > 2400 inicia turbulencia en el centro
- Re > 4000 flujo turbulento
- Re< 2100 flujo laminar

### En láminas, si l es el distanciamiento entre ellas,
- Re > 1000 turbulento

### En canales, con profundidad l
- Re > 500 turbulento

### En esferas, l diámetro de la esfera
- Re > 1 turbulento

# Presión vs diferencia de altura

Balance de fuerza en volumen de control de un fluido estático:



$$
\begin{aligned}
\Sigma F &= 0 \rightarrow \text{Esta en equilibrio} \\[10pt]
F_2 - F_1 - m_\text{fluido}g &= 0 \: \: ; \: \: P = \frac{F}{A} \: \: ; \: \: m = \rho V = -\rho A \mathrm{d}h \\[10pt]
P_2\cancel{A} - P_1\cancel{A}& = -\rho g \cancel{A} \mathrm{d}h \\[10pt]
\mathrm{d}P &= -\rho g \mathrm{d}h \\[10pt]
\frac{\mathrm{d}P}{\mathrm{d}h} &= -\rho g  \\[10pt]
\end{aligned}
$$

Integramos:

$$
\begin{aligned}
\int_{P_1}^{P_2} \mathrm{d}P &= -\rho g \int_{h_1}^{h_2} \mathrm{d}h  \\[10pt]
P_2 - P_1 &= \rho g ({h_2}-{h_1}) \rightarrow h_2 \: \text{esta debajo del marco de referencia en sentido contrario el eje de cordenada Y} \\[10pt]
P_2 + \rho g h_2 &= P_1 \rho g h_1 = \Phi  \\[10pt]
\end{aligned}
$$


# Hidrómetro

Al usarlo en agua, no se evidencia un cambio de altura, por tanto $W_\text{hidrometro} = \gamma_{\text{agua}}V$

pero en el fluido x si tendra una diferencia de altura entre altura inicial de y altura final de el fluido al sumergir el hidrometro: $W_\text{hidrometro} = \gamma_{\text{x}}(V - A\Delta h)$

igualando ambas expresiones anteriores:

$$
\begin{aligned}
\Delta h &= \frac{V}{A}(1-\frac{\gamma_{\text{agua}}}{\gamma_{\text{x}}}) \: \: ; \: \: \text{con} S_x = \frac{\gamma_{\text{x}}}{\gamma_{\text{agua}}}\\[10pt]
\Delta h &= \frac{V}{A}(1-\frac{1}{S_x})\\[10pt]
\end{aligned}
$$
