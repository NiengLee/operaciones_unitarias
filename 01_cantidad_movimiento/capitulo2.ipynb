# Ecuación de Continuidad o Conservación de la masa :

$$
\begin{aligned}
\frac{\mathrm{d}M_{\text{sistema}}}{\mathrm{d}t} &= \Sigma_i (\dot{m}_i)_{\text[in]} - \Sigma_i (\dot{m}_j)_{\text[out]}\\[10pt]
0 &= \Sigma_i (\dot{m}_i)_{\text{in}} - \Sigma_i (\dot{m}_j)_{\text{out}} \rightarrow \text{Para régimen estacionario}\\[10pt]
\dot{m} &= \rho A \vec{V}  = \text{constante}\\[10pt]
\checkmark \left(\rho A \vec{V}\right)_{\text{in}} &= \left(\rho A \vec{V}\right)_{\text{out}}\\[10pt]
G &= \frac{\dot{m}}{A}\\[10pt]
\end{aligned}
$$

- Ecuación de Euler en su forma general.
- Ecuación de Bernoulli para Fluidos Incompresibles

Asumimos flujo estacionario:

$$
\begin{aligned}
\Sigma_{\text{in}}m_ie_i - \Sigma_{\text{out}}m_je_j + Q - W &= \cancel{\frac{\mathrm{d}E}{\mathrm{d}t}}\\[10pt]
e_n = u_n + \frac{\vec V^2_n}{2} + gZ_n + P_nv_n &\rightarrow \text{Donde e es la contribución de energía en todas sus formas}\\[10pt]
\Sigma_{\text{in}}m_i\left(u_i + \frac{\vec V^2_i}{2} + gZ_i + P_iv_i\right) - \Sigma_{\text{out}}m_j\left(u_j + \frac{\vec V^2_j}{2} + gZ_j + P_jv_j\right) + Q - W &= 0\\[10pt]
u_i = u_j &\rightarrow \text{no hay cambio de temperatura significativa, por lo tanto, la energía interna es constante}\\[10pt]
\left(\frac{\vec V^2_i}{2} + gZ_i + P_iv_i\right) - \left(\frac{\vec V^2_j}{2} + gZ_j + P_jv_j\right) + q - w = 0 &\rightarrow \text{mi=mj por flujo en tuberia, la masa es constante, y haremos análisis por kg}\\[10pt]
\frac{\vec V^2_i}{2} + gZ_i + \frac{P_i}{\rho_i} - \frac{\vec V^2_j}{2} - gZ_j - \frac{P_j}{\rho_i} + \cancel{q} - \cancel{w} &= 0 \: \: ; \: \: \rho = \frac{1}{v} \\[10pt]
\end{aligned}
$$

Se anula w y q, porque se consideran procesos adiabáticos y no hay inyección o generacíon de trabajo mecánico. Finalmente agregamos perdidas de energía por flujo $\mathrm{d}h_{f_{\text{e-s}}} = \frac{L\vec{v}^2}{D2}f$

$$
\checkmark \frac{P_i}{\rho_i} + \frac{\vec V^2_i}{2} + gZ_i - \frac{P_j}{\rho_i} - \frac{\vec V^2_j}{2} - gZ_j - \frac{L\vec{v}^2}{D2}f_D = 0
$$



# Ecuación de Energía o “Balance” total de energía :

Asumimos flujo estacionario:


$$
\begin{aligned}
\Sigma_{\text{in}}m_i\left(u_i + \frac{\vec V^2_i}{2} + gZ_i + P_iv_i\right) - \Sigma_{\text{out}}m_j\left(u_j + \frac{\vec V^2_j}{2} + gZ_j + P_jv_j\right) + Q - W &= \cancel{\frac{\mathrm{d}E}{\mathrm{d}t}}\\[10pt]
u_i \neq u_j &\rightarrow \text{hay cambio de temperatura}\\[10pt]
h_i &= u_i + P_iv_i\\[10pt]
\left(h_i + \frac{\vec V^2_i}{2} + gZ_i\right) - \left(h_j + \frac{\vec V^2_j}{2} + gZ_j\right) + q - w = 0 &\rightarrow \text{por flujo en tuberia, la masa es constante, y haremos análisis por kg}\\[10pt]
h_i + \frac{\vec V^2_i}{2} + gZ_i - h_j - \frac{\vec V^2_j}{2} - gZ_j + q &= 0
\end{aligned}
$$

Recordemos que $\Delta h = \int_{T_\text{in}}^{T_\text{out}} C_p(T)\mathrm{d}T$
$$
\checkmark-\int_{T_\text{in}}^{T_\text{out}} C_p(T)\mathrm{d}T + \frac{\vec V^2_i}{2} + gZ_i  - \frac{\vec V^2_j}{2} - gZ_j + q = 0
$$

# Detalles de las pérdidas por fricción:

Las pérdidas por fricción se formulan como:

$$
\begin{aligned}
h_f = 4 \frac{l}{D} \frac{\vec{v}^2}{2g} f &= \frac{l}{D} \frac{\vec{v}^2}{2g} f_{\text{Darcy}}\\[10pt]
f_{\text{Darcy}} &= f_{\text{Moody}} = 4f
\end{aligned}
$$

Definiciones:

- $h_f$ : Pérdida de carga por fricción. Es la energía que se pierde por unidad de peso del fluido. Se mide en metros de columna de fluido. NO CONFUNDIR CON $h_{f_{\text{e-s}}}$, ya que en esta ultima no se incluye el termino de la gravedad!
- l : Longitud de tubería.
- D : Diámetro interno de la tubería.
- $\frac{L}{D}$ : Relación geométrica que indica cuántos “diámetros” caben en la longitud total. Entre más larga la tubería, mayor pérdida.
- $\vec{v}$ : Velocidad media del fluido.
- g : gravedad
- f : Factor de fricción de Fanning.


 f es un factor que definió J.T. Fanning en 1877 en su libro: “A practical treatise on hydraulic and water supply engineering”. Van Nostrand, New York. Hay una equivalencia posterior

```python
def h_perdida_friccion(l : float,D: float,vel: float,f_d: float,g: float=9.8) -> float: 
    """
    Datos de entrada:
    - longitud tubería : l [=] m
    - diametro de tubería : D [=] m
    - velocidad de flujo : vel [=] m/s
    - factor Darcy : f_d [=] m/s
    - gravedad : g [=] m/s², = 9.8
    
    Datos de salida:
    - Distancia que no pudo alcanzar por la fricción : h [=] m
    """
    
    h_f_es_perdida = (l*(vel**2)*f_d)/(2*D)
    h_f_perdida = h_f_es_perdida/g
    
    return {
            "h_f(e-s) de perdidas" : h_f_es_perdida,
            "h_f de perdidas sobre g" : h_f_perdida,
        } 
```

# Cálculo de $f_{Darcy}$

## Para régimen laminar:

$$
f = \frac{16}{Re}
$$
$$
f_{\text{Darcy}} = \frac{64}{Re}
$$

## Para régimen Turbulento:
Esta una opción que es usar el diagrama de Moody, o usar la expresión de Shacham:

$La **correlación explícita de Shacham** para el **factor de fricción de Darcy** (flujo turbulento en tuberías) es:

$$
f_D = \left[-2 \log\left(\frac{\frac{\epsilon}{D}}{3.71}-\frac{5.02}{\mathrm{Re}}\log\left(\frac{\frac{\epsilon}{D}}{3.71}+\frac{14.5}{\mathrm{Re}}\right)\right)\right]^{-2}
$$

Términos:

- $f_D$  : factor de fricción de **Darcy**
- $\epsilon$ : rugosidad absoluta de la tubería
- D : diámetro interno
- $\mathrm{Re}$ : número de Reynolds



Es una aproximación explícita de la ecuación implícita de Colebrook–White, válida en régimen turbulento.


```python
def f_darcy_laminar(Re: float) -> dict :
    """
    Calcula el factor de fricción de Darcy para flujo laminar.

    Datos de entrada:
    - Re : Número de Reynolds (adimensional). Válido en régimen turbulento (Re > 4000).

    Retorna
    - f : Factor de fricción de Darcy (adimensional).
    """
    
    if Re >= 2000:
        return {
            "message" : "Reynolds > 2000, flujo no laminar",
            "Factor Darcy" : f"este cálculo no aplica para Reynolds = {Re}"
        }
            
    f = 64/Re
    
    return {
        "message" : "Reynolds > 4000, flujo turbulento",
        "Factor Darcy" : f
    }


def f_darcy_shacham(Re: float, epsilon: float, D: float) -> dict :
    """
    Calcula el factor de fricción de Darcy usando la correlación explícita de Shacham.

    Datos de entrada:
    - Re : Número de Reynolds (adimensional). Válido en régimen turbulento (Re > 4000).
    - epsilon : Rugosidad absoluta de la tubería [=]m.
    - D : Diámetro interno de la tubería [=] m.

    Retorna
    - f : Factor de fricción de Darcy (adimensional).
    """

    if Re < 2000 :
        return {
            "message" : "Reynolds es < 2000, se asume flujo laminar",
            "Factor Darcy" : f_darcy_laminar(Re)
        }

    eps_D = epsilon / D

    termino_1 = eps_D / 3.7
    termino_2 = 5.02 / Re
    inner_log = termino_1 + (14.5 / Re)

    argument = termino_1 - termino_2 * math.log10(inner_log)

    f = (-2.0 * math.log10(argument)) ** -2

    if Re < 4000 :
        return {
            "message" : "Reynolds no es > 4000, Reynolds no es < 2000, esta en transición entre laminar y turbulento - Use el factor con precaución",
            "Factor Darcy" : f
        }
    else :
        return {
            "message" : "Reynolds > 4000, flujo turbulento",
            "Factor Darcy" : f
        }
```

# Calculo de pérdidas menores en accesorios

## Método 1 K:
$$
h_f = \frac{l}{D} \frac{\vec{v}^2}{2g} f_{\text{Darcy}} = K · \frac{\vec{v}^2}{2g} = h_{\text{accesorio}}
$$

- K agrupa $\frac{l}{D} f_{\text{Darcy}}$.
- K es un valor constante tomado de tablas. 
- Este metodo de solo tomar 1 K ya no se usa por los avances tecnológicos, que ofrecen mayor precisión.

## Método 2 K (Hooper):

$$
K = \frac{K_1}{Re} + K_{\infty} (1 + \frac{1}{D_i})
$$

- $\frac{K_1}{Re}$ : Captura el efecto del número de Reynolds. Hace que la pérdida dependa del régimen de flujo.
- $K_{\infty}$ : es el valor de K cuando Re → ∞ (flujo completamente turbulento).
- $D_i$ : Diámetro interno aproximado del accesorio (tener cuidado con las unidades, muchas veces es en pulgadas).

La K resultante se reemplaza luego en $h_{\text{accesorio}} = K \frac{\vec{v}^2}{2g}$

## Método 3 K (Darby):

$$
K = \frac{K_1}{Re} + K_i (1 + \frac{K_d}{D_i^{0.3}})
$$

- Las K son valores tomados de referencia bibliográficas, manuales o catálogos de accesorios.
- $D_i$ : Diámetro interno aproximado del accesorio (tener cuidado con las unidades, muchas veces es en pulgadas).

La K resultante se reemplaza luego en $h_{\text{accesorio}} = K \frac{\vec{v}^2}{2g}$

```python
def h_perdidas_2k(K_1 : float, K_inf : float, D_i : float, Re : float, vel:float, g : float = 9.8) -> dict :
    """
    Calcula las perdidas de energía en accesorios (Hooper).

    Datos de entrada:
    - K_1 : Tomado de datos documentados y manuales
    - K_inf : Tomado de datos documentados y manuales
    - D_i : Diámetro interno de la tubería [=] m.
    - Re : Número de Reynolds [=] adimensional.
    - vel : velocidad del fluido que pasa por el accesorio [=] m/s.
    - g : gravedad [=] m/s, default = 9.8.

    Retorna
    - f : Factor de fricción de Darcy (adimensional).
    """
    D_i = D_i * 39.3701 # inches
    K = (K_1/Re) + (K_inf * (1 + (1/D_i)))
    h_f_es_perdida = K * ((vel**2)/2)
    h_f_perdida = h_f_es_perdida/g
    
    return {
            "h_f(e-s) de perdidas" : h_f_es_perdida,
            "h_f de perdidas sobre g" : h_f_perdida,
        } 


def h_perdidas_3k(K_1 : float, K_i : float, K_d : float, D_i : float, Re : float, vel:float, g : float = 9.8) -> float :
    """
    Calcula las perdidas de energía en accesorios (Darby).

    Datos de entrada:
    - K_1 : Tomado de datos documentados y manuales
    - K_i : Tomado de datos documentados y manuales
    - K_d : Tomado de datos documentados y manuales
    - D_i : Diámetro interno de la tubería [=] m.
    - Re : Número de Reynolds [=] adimensional.
    - vel : velocidad del fluido que pasa por el accesorio [=] m/s.
    - g : gravedad [=] m/s, default = 9.8.

    Retorna
    - f : Factor de fricción de Darcy (adimensional).
    """
    D_i = D_i * 39.3701 # inches
    K = (K_1/Re) + (K_i * (1 + (K_d/(D_i**0.3))))
    h_f_es_perdida = K * ((vel**2)/2)
    h_f_perdida = h_f_es_perdida/g
    
    return {
            "h_f(e-s) de perdidas" : h_f_es_perdida,
            "h_f de perdidas sobre g" : h_f_perdida,
        } 
```

# Modificaciónes al balance de energía cuando hay consumo de trabajo:



Volvemos a la ecuación de balance de energía mecánica de flujo, y asumiremos lo siguiente:

- tenemos un líquido fluyendo en la figura
- la densidad no cambia para fluidos líquidos
- masa en el punto 1 es igual a la masa del punto 2
- no hay cambio de temperatura (al menos no significativamente, y por tanto, no hay cambio de energía interna)
- el proceso es adiabático, no hay perdidas significativas de calor
- la bomba ingresa energía para poder llevar el flujo del punto 1 al punto 2
- estado estacionario.

$$
\begin{aligned}
\frac{P_1}{\rho_1} + \frac{\vec V^2_1}{2} + gZ_1 - \frac{P_2}{\rho_2} - \frac{\vec V^2_2}{2} - gZ_2 + \cancel{q} + w - h_{f_\text{e-s}} &= 0 
\end{aligned}
$$

Como pueden ver, $h_f$ es un factor encargado de consumo de energía por procesos irreversibles, por eso el signo negativo. despejando el trabajo necesario de la bomba para llevar a cabo el proceso:

$$
\begin{aligned}
w &= \frac{P_2}{\rho_2} + \frac{\vec V^2_2}{2} + gZ_2 - \frac{P_1}{\rho_1} - \frac{\vec V^2_1}{2} - gZ_1 + h_{f_\text{e-s}} \\[10pt]
\checkmark w &= \frac{\Delta P}{\rho} + \frac{\vec V_2^1 - \vec V_1^2}{2} + g\Delta Z + h_{f_\text{e-s}} \\[10pt]
\frac{w}{g} &= \frac{\Delta P}{\rho g} + \frac{\vec V_2^1 - \vec V_1^2}{2g} + \Delta Z + \frac{h_{f_\text{e-s}}}{g} \\[10pt]
\checkmark \frac{w}{g} &= \frac{\Delta P}{\rho g} + \frac{\vec V_2^1 - \vec V_1^2}{2g} + \Delta Z + h_f \\[10pt]
\text{con} \: h_f &= h_1 + h_2 + h_3 + h_4 + \text{los tramos de tuberias que no tienen accesorios (6 secciones)}
\end{aligned}
$$

Se recomienda tener atención sobre la letra h, puede ser entalpía, perdidas por fricción de flujo, altura. Evitar confusiones con esta variable.

Por lo tanto, el trabajo de la bomba tiene que ser como mínimo equivalente a la energía del cambio de presión, cambio de energía cinética, cambio de energía potencial y superar las perdidas por fricción de flujo en tuberias y accesorios para que el proceso pueda realizarse.

Se debe tener cuidado con la expresión de la gravedad, ya que $h_f$ la tiene involucrada y puede afectar la ecuación de balance de energía. Cabe recordar entonces:

$$
h_f = \frac{l}{D} \frac{\vec{v}^2}{2g} f_{\text{Darcy}} = K · \frac{\vec{v}^2}{2g} = h_{\text{accesorio}} \: \: ; \: \: h_{f_{\text{e-s}}} = \frac{l\vec{v}^2}{D2}f_{\text{Darcy}}\\[10pt]
$$
$$
h_f = \frac{h_{f_{\text{e-s}}}}{g}
$$

Usesé la expresión mas conveniente, con precaución y dependiendo del contexto.

# Ecuación de Bernoulli:

Tenemos del balance de energía mecánica para procesos estacionarios, adiabáticos :

$$
\frac{P_1}{\rho_1} + \frac{\vec V^2_1}{2} + gZ_1 - \frac{P_2}{\rho_2} - \frac{\vec V^2_2}{2} - gZ_2 + \cancel{q} + w - h_{f_\text{e-s}} = 0 
$$

Reordenamos :

$$
\frac{P_1}{\rho_1} + \frac{\vec V^2_1}{2} + gZ_1 + w  = \frac{P_2}{\rho_2} + \frac{\vec V^2_2}{2} + gZ_2 + h_{f_\text{e-s}} \rightarrow \text{Ecuación de Bernoulli} \: \checkmark
$$
