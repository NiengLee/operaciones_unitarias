Si la presin del aire en una tubera es de 450 kPa g, Qu lectura H, habr en un manmetro de tubo en U que contiene mercurio? Use h = 1,5 cm

- b) Incluya el peso de la columna de aire, suponiendo que Taire = 20 C y calcule el porcentaje de error de la parte a. Rta//3.36 m, % error= 4 x 10-4



```python
print("Realizado por ExergiaCode: ")
h = 1.5 * (1/100) # m
Pair = 450 * (1000) # Pa
g = 9.81 # m /s
den_Hg = 13600  # kg / m
den_aire = 1.2 # kg/m

H = (Pair) / (den_Hg * g)
print(f"la distancia H es : {H:,.3f} m")

H_aire = (Pair + (den_aire *g*h)) / (den_Hg * g)
print(f"la distancia H es : {H_aire:,.3f} m")

error = abs(H_aire - H) / H_aire
print(f"error = {error*100:,.8f} %")
```

El hidrmetro mostrado en la figura sin cobre tiene una masa de 0,01 kg. Est diseado para flotar a la mitad del vstago de 12 cm de longitud en agua pura. 

- a) Calcular la masa de cobre requerida. Rta// 0.0177 Kg 
- b) Calcular la densidad mnima del lquido que puede ser leda. Rta// 957 Kg/m3 
- c) Calcular la densidad mxima del lquido puede ser leda. Rta// 1046Kg/m3





```python
print("Realizado por ExergiaCode: ")
m_hidrometro = 0.01 # kg
g = 9.8 # m/s
den_agua = 1000 # kg/m
D_vastago = 5 * (1/1000) # m
h_vastagoT = 12 * (1/100) # m
h_vastago = h_vastagoT/2 # m
D_cilindro = 1.5 * (1/100) # m
h_cilindro = 15 * (1/100) # m

r_cilindro = D_cilindro/2 # m
r_vastago = D_vastago/2 # m

m_cobre = (den_agua * ((np.pi * (r_cilindro**2) * h_cilindro) + (np.pi * (r_vastago**2) * h_vastago))) - m_hidrometro

print(f"la masa de cobre es: {m_cobre:,.5f} kg")

m_total = m_cobre + m_hidrometro

# La densidad mnima ocurre cuando el hidrmetro se hunde lo mximo posible, es decir, cuando el vstago est completamente sumergido
# (12 cm adicionales respecto a la posicin media).

den_min = (m_total)/((np.pi * (r_cilindro**2) * h_cilindro) + (np.pi * (r_vastago**2) * h_vastagoT))
print(f"La densidad mnima es: {den_min:,.3f} kg/m")

# La densidad mxima ocurre cuando el vstago queda por completo sobre el nivel del fluido, es decir, cuando el vstago est completamente fuera del fluido.
# (0 cm del vastago esta sumergido).

h_vastago_en_aire = (1/1000) * h_vastagoT
den_max = (m_total)/((np.pi * (r_cilindro**2) * h_cilindro) + (np.pi * (r_vastago**2) * h_vastago_en_aire))
print(f"La densidad mxima es: {den_max:,.3f} kg/m")
```

# Fuerza de empuje
Una corona tiene una masa de 2,50 kg. Se sabe que:

- La densidad del oro es _oro = 19 300 kg/m
- La densidad de la plata es _plata = 10 500 kg/m
- La densidad del agua es _agua = 1 000 kg/m
- La aceleracin gravitacional es g = 9,81 m/s

Al medir su peso:

- En aire, la balanza marca 24,525 N
- Sumergida completamente en agua, la balanza marca 22,965 N

...

- Calcule la fuerza de empuje que acta sobre la corona.
- Determine el volumen de la corona a partir del empuje.
- Calcule la densidad experimental de la corona.
- La corona es de oro puro?.

## Planteamiento





```python
print("Realizado por ExergiaCode: ")

#   Informacin inicial:
den_oro = 19300 # kg/m
den_plata = 10500 # kg/m
den_agua = 1000 # kg/m
g = 9.81 #m/s

W_medido_aire = 24.525 # N
W_medido_agua = 22.965 # N

#   Objetivo : Calcular la fuerza de empuje sobre la corona.
F_empuje = W_medido_aire - W_medido_agua
print(f"La fuerza de empuje es : {F_empuje:,.3f} N")

# F_empuje = volumen agua desplazado * densidad agua * g
# F_empuje / (densidad agua * g) = volumen agua desplazado
V_agua_desplazado = F_empuje/(den_agua * g)
print(f"El volumen desplazado es : {V_agua_desplazado:,.8f} m")

V_corona = V_agua_desplazado
# si la corona es es 100% oro, entonces:
m_corona_oro = den_oro * V_corona 
print(f"la masa de corona en 100% es : {m_corona_oro:,.5f} kg") # 3.06911kg

# 3.06911kg no concuerda con la masa inicial de 2.5

# Sistema de 2x2
# masa_plata + masa_oro = 2.5
# masa_plata/den_plata + masa_oro/den_oro = V_corona
m_o = (V_corona - (2.5/den_plata))/((1/den_oro)-(1/den_plata))
m_p = 2.5 - m_o
comp_o = m_o/2.5
comp_p = m_p/2.5

print(f"masa de oro = {m_o:,.3f}kg ; masa de plata = {m_p:,.3f}kg ; total = {m_o+m_p}kg")
print(f"composicion oro = {comp_o:,.5f} % ; composicion plata = {comp_p:,.5f} %")
print(f"Densidad experimental de la corona = {(m_o+m_p)/(V_corona):,.3f} kg/m")
```

- Un barco flota porque su densidad promedio total (estructura + carga + aire en su interior) es menor que la densidad del agua.
